```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
boston_df=pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv', sep =",")
boston_df.isnull().sum()

```

```python
plt.figure(figsize=(12,8))
sns.distplot(boston_df['medv'], hist = True)
plt.axvline(boston_df['medv'].mean(), linestyle='-')
plt.show()

```

```python
plt.figure(figsize=(12,8))
sns.heatmap(boston_df.corr(method='pearson'), annot =True, cmap='RdPu')
plt.show()

```

```python
plt.figure(figsize=(6,6))
plt.scatter(boston_df['rm'], boston_df['medv'])
plt.show()

```

```python
#correlation
plt.figure(figsize=(12,8))
sns.heatmap(boston_df.corr(method='pearson'), annot =True, cmap='RdPu')
plt.show()

```

```python
plt.figure(figsize=(12,8))
plt.scatter(boston_df['rm'], boston_df['medv'])
plt.show()

```

```python
#data preprocessing
from sklearn.preprocessing import MinMaxScaler
Scaler =MinMaxScaler(feature_range = (0, 1))

```

```python
X= boston_df.drop(['rad', 'medv'], axis =1)
Y = boston_df['medv']

```

```python
X_Scaled = Scaler.fit_transform(X)

```

```python
#Linear Regression
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, Y_train, Y_test= train_test_split(X_Scaled, Y, test_size = 0.2 , random_state = 3)

from sklearn.linear_model import LinearRegression
lin_model= LinearRegression()
```

```python

lin_model.fit(X_train, Y_train)

# Predictions
y_pred = lin_model.predict(X_test)

# Calculate MAE, MSE, and RMSE
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(Y_test, y_pred)
mse = mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
